```python
# Change in Market Cap Growth by CEO

data_to_load = Path(f"Resources/Queried/MCGbyCEO.csv")

data_df = pd.read_csv(data_to_load)

data_df


CompanyList = data_df.CompanyName.unique()
for company in CompanyList:

    #Filter down to correct company
    company_data = data_df[(data_df.CompanyName == company)]

    #create list of ceos 
    CeoList = company_data.CeoName.unique()

    #create dictionary of dataframes, each one holding one ceo's tenure
    ceo_dict =dict()
    fig, ax = plt.subplots()
    for name in CeoList:
        ceo_dict[name] = company_data[(company_data.CeoName == name)]
        ax.plot("Year", "Change", data=ceo_dict[name], label = name )
    
    ax.set(xlabel="Year", ylabel="Change in MarketCap", title= f"MarketCap Growth by CEO - {company}")
    ax.legend()

    plt.show
```

```python
# Change in Market Cap Growth by CEO

data_to_load = Path(f"Resources/Queried/MCbyCEO.csv")

data_df = pd.read_csv(data_to_load)

data_df


CompanyList = data_df.CompanyName.unique()

for company in CompanyList:

    

    #Filter down to correct company
    company_data = data_df[(data_df.CompanyName == company)]

    #create list of ceos 
    CeoList = company_data.CeoName.unique()

    #create dictionary of dataframes, each one holding one ceo's tenure
    ceo_dict =dict()
    fig, ax = plt.subplots()
    for name in CeoList:
        ceo_dict[name] = company_data[(company_data.CeoName == name)]
        ax.plot("Year", "MarketCap(Billions)", data=ceo_dict[name], label = name )
    
    ax.set(xlabel="Year", ylabel="MarketCap", title= f"MarketCap by CEO - {company}")
    ax.legend()

    plt.show
```

```python
# Change in Market Cap by Company

data_to_load = Path(f"Resources/Queried/MCbyCEO.csv")

data_df = pd.read_csv(data_to_load)

data_df


CompanyList = data_df.CompanyName.unique()
company_dict =dict()
fig, ax = plt.subplots()

for company in CompanyList:
    
    #create dictionary of dataframes, each one holding one ceo's tenure
    company_dict[company] = data_df[(data_df.CompanyName == company)]
    ax.plot("Year", "MarketCap(Billions)", data =company_dict[company], label = company )
    
    

ax.set(xlabel="Year", ylabel="MarketCap", title= f"MarketCap by Company")

ax.legend()
plt.show
```
