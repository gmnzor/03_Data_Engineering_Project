```python
import numpy as np
import pandas as pd
import datetime as dt
from pathlib import Path
```

```python
# Initialize list containing company name
# Order in this list will determine CompanyID
CompanyNames = ["Amazon", "Apple", "Facebook", "Google", "Netflix"]

#initialize dataframe that will hold all stock data
stock_df = pd.DataFrame()
```

```python
# For loop to pull company name, read in csv file, and create a 
# master list of stock data for all companies combined
for company in CompanyNames:

    # defines path where individual companies' stock data can be found
    data_to_load = Path(f"Resources/{company}.csv")

    # reads in stock data and assigns it to dataframe
    company_data = pd.read_csv(data_to_load)

    # creates new column and assigns companyID relative to position 
    # of company in original companyNames list
    company_data.insert(0,"CompanyID", CompanyNames.index(company), True)
    
    # adds individual stock data to master list of stock information
    stock_df = pd.concat([stock_df,company_data],ignore_index=True, sort=False)

stock_df

```
