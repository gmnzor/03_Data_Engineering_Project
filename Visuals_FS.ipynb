```python
import numpy as np
import pandas as pd
import datetime as dt
from pathlib import Path
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
data_to_load = Path("Resources/Queried/MCGbyCEO.csv")
data_df = pd.read_csv(data_to_load)

print(data_df.head())
```

```python
# Change in Market Cap Growth by CEO
def plot_market_cap_growth_by_ceo(data_df):
    CompanyList = data_df.CompanyName.unique()
    for company in CompanyList:
        # Filter down to correct company
        company_data = data_df[data_df.CompanyName == company]

        # Create list of CEOs 
        CeoList = company_data.CeoName.unique()

        # Create dictionary of dataframes, each one holding one CEO's tenure
        ceo_dict = {}
        fig, ax = plt.subplots(figsize=(10, 6))
        for name in CeoList:
            ceo_dict[name] = company_data[company_data.CeoName == name]
            ax.plot(ceo_dict[name]["Year"], ceo_dict[name]["Change"], label=name)

        ax.set_xlabel("Year")
        ax.set_ylabel("Change in Market Cap")
        ax.set_title(f"Market Cap Growth by CEO - {company}")
        ax.legend()
        ax.grid(True)
        plt.show()

plot_market_cap_growth_by_ceo(data_df)
```

```python

# Change in Market Cap by Company
def plot_market_cap_by_company(data_df):
    # Filter data from 2012 onwards
    data_df = data_df[data_df['Year'] >= 2012]
    
    CompanyList = data_df.CompanyName.unique()
    company_dict = {}
    fig, ax = plt.subplots(figsize=(10, 6))

    for company in CompanyList:
        # Create dictionary of dataframes, each one holding one company's data
        company_dict[company] = data_df[data_df.CompanyName == company]
        ax.plot(company_dict[company]["Year"], company_dict[company]["MarketCap(Billions)"], label=company)

    ax.set_xlabel("Year")
    ax.set_ylabel("Market Cap (Billions)")
    ax.set_title("Market Cap by Company (from 2012)")
    ax.legend()
    ax.grid(True)
    plt.show()

plot_market_cap_by_company(data_df)
```
