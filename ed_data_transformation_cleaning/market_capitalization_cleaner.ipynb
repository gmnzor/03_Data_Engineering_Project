```python
import pandas as pd
from pathlib import Path
```

```python
# Initialize company names
CompanyNames = ["Amazon", "Apple", "Facebook", "Google", "Netflix"]

# Initialize dataframe that will hold all NET WORTH information (market capitalization)
market_cap = pd.DataFrame()
```

```python
#For loop make a master key of market cap data for all companies

for company in CompanyNames:

    # read in data from all companies
    company_cap_data = pd.read_excel(f"../Resources/{company}_market_capitalization.xlsx")
    
    # creates new column and assigns companyID relative to position 
    # of company in original companyNames list
    company_cap_data.insert(0,"CompanyID", CompanyNames.index(company), True)

     # adds individual market capitalization data to master list of NET WORTH information
    market_cap = pd.concat([market_cap,company_cap_data],ignore_index=True, sort=False)

market_cap
```

```python
market_cap["MarketCap"]= market_cap["MarketCap"].str.replace("$", "")
market_cap
```

```python
market_cap.insert(0,"CBilTrill", "1", True)
market_cap.insert(0,"MarketCap(Billions)", "0", True)
market_cap
```

```python
market_cap.loc[market_cap['MarketCap'].str.split(expand=True)[1] == "T", 'CBilTrill'] = 1000
market_cap
```

```python
market_cap["MarketCap"]= market_cap["MarketCap"].str.replace(" T", "")
market_cap["MarketCap"]= market_cap["MarketCap"].str.replace(" B", "")
market_cap
```

```python
market_cap["MarketCap"] = market_cap["MarketCap"].astype("float")
market_cap["MarketCap(Billions)"]=market_cap["MarketCap(Billions)"].astype("float")
market_cap["CBilTrill"] = market_cap["CBilTrill"].astype("int")

```

```python
market_cap.dtypes
```

```python
market_cap["MarketCap(Billions)"] = market_cap["MarketCap"] * market_cap["CBilTrill"]
market_cap
```

```python
market_cap = market_cap.drop(["CBilTrill","MarketCap"], axis=1)
```

```python
# write stock data to CSVfile
market_cap.to_csv("../Resources/market_cap.csv", index=False)
```
